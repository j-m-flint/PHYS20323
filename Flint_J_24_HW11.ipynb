```python
import math
import numpy as np

#PART 1
class Particle(object):
    def __init__(self, x, y, z, u, v, w, m):
        self.xposition = x
        self.yposition = y
        self.zposition = z
        self.uvelocity = u
        self.vvelocity = v
        self.wvelocity = w
        self.mass = m
     
    # PART 2   
    def kinetic_energy(self):
        velocity = math.sqrt(self.uvelocity**2 + self.vvelocity**2 + self.wvelocity**2)
        KE = 0.5*self.mass*velocity**2
        print("kinetic energy=", KE)
     
    #PART 3   
    def particle_distance(self, other):
        dx = other.xposition - self.xposition
        dy = other.yposition - self.yposition
        dz = other.zposition - self.zposition
        distance = math.sqrt(dx**2 + dy**2 + dz**2)
        return distance
        print("distance=", distance)
     
    #PART 4   
    def update_position(self, dt):
        self.xposition += self.uvelocity * dt
        self.yposition += self.vvelocity * dt
        self.zposition += self.wvelocity * dt
        return self.xposition, self.yposition, self.zposition
        print("Updated position:", "(", self.xposition, self.yposition, self.zposition,")")

#PART 5        
class ChargedParticle(Particle):

    def __init__(self, mass, x, y, z, u, v, w, charge):
        self.mass = mass
        self.xposition = x
        self.yposition = y
        self.zposition = z
        self.uvelocity = u
        self.vvelocity = v
        self.wvelocity = w
        self.charge = charge
        
#EXCERCISE 2_______________________________________________________________________       

#PART 6
    def find_separation(self, other):
        return self.particle_distance(other)

    def interact(self, other):
        separation = self.find_separation(other)
        if separation < 0.25:  
            print("Interaction")
            if self.charge * other.charge < 0:  # Opposite 
                print("Merge")
                self.uvelocity, self.vvelocity, self.wvelocity = 0, 0, 0
                other.uvelocity, other.vvelocity, other.wvelocity = 0, 0, 0
                self.charge, other.charge = 0, 0
            else:  # Same sign
                print("Repel")
                factor_self = -1 * (self.charge + other.charge) * (self.mass / other.mass)
                factor_other = -1 * (self.charge + other.charge) * (other.mass / self.mass)

                self.uvelocity *= factor_self
                self.vvelocity *= factor_self
                self.wvelocity *= factor_self

                other.uvelocity *= factor_other
                other.vvelocity *= factor_other
                other.wvelocity *= factor_other

#PART 7/8
def simulate_interaction(p1, p2, dt, steps):
    for step in range(steps):
        p1.update_position(dt)
        p2.update_position(dt)
        
        p1.interact(p2)
        
       
        print(f"Step {step + 1}:")
        print(f"  P1 Position: ({p1.xposition:.2f}, {p1.yposition:.2f}, {p1.zposition:.2f}) "
              f"Velocity: ({p1.uvelocity:.2f}, {p1.vvelocity:.2f}, {p1.wvelocity:.2f})")
        print(f"  P2 Position: ({p2.xposition:.2f}, {p2.yposition:.2f}, {p2.zposition:.2f}) "
              f"Velocity: ({p2.uvelocity:.2f}, {p2.vvelocity:.2f}, {p2.wvelocity:.2f})")
        print("--------------------------------------------------------------")

#PART 7
p1 = ChargedParticle(-5, -5, -5, 1, 1, 1, 1, 0.5)
p2 = ChargedParticle(5, 5, 5, -1, -1, -1, 1, -0.5)
simulate_interaction(p1, p2, dt=0.05, steps=300)

#PART 8
p1 = ChargedParticle(-5, -5, -5, 2, 2, 2, 1, 0.5)
p2 = ChargedParticle(5, 5, 5, -1, -1, -1, 1, 2.0)
simulate_interaction(p1, p2, dt=0.05, steps=300)
```
